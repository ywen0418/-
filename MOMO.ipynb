```python
import requests
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'}
momo = requests.get('https://www.momoshop.com.tw/category/DgrpCategory.jsp?d_code=4300100018&TOP30=Y',headers=headers)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(momo.text,'lxml')
soup
```

```python
soup_eachgood = soup.select('.eachGood')
len(soup_eachgood)
```
